## 计算 事件数-执行时间 测试结果

```python
# 计算事件数-执行时间
import os
import time
import pandas as pd
import warnings
import re  # 导入正则表达式模块
from ocpa.objects.log.importer.csv import factory as ocel_import_factory
from ocpa.algo.discovery.ocpn import algorithm as ocpn_discovery_factory

# 忽略警告信息
warnings.filterwarnings("ignore")

# 输入和输出目录
input_dir = r"C:\Users\lenovo\Desktop\毕业论文\案例分析\日志\变体0-9子日志\sublog-递增变体"
output_file = r"C:\Users\lenovo\Desktop\毕业论文\案例分析\日志\变体0-9子日志\递增变体-事件数-时间.csv"

# 初始化输出数据
output_data = []

# 递归遍历输入目录中的每个文件
for root, dirs, files in os.walk(input_dir):
    for file in files:
        if file.endswith(".csv"):  # 检查文件是否为 CSV 格式
            file_path = os.path.join(root, file)
            
            # 从文件名中提取事件数
            # 假设文件名格式为 "9660E_2415UO_12880Oevent_variant_[0].csv"
            # 使用正则表达式提取文件名中的第一个数字
            match = re.search(r'\d+', file)
            if match:
                num_events = int(match.group())
            else:
                print(f"Error extracting num_events from file name {file}. Skipping this file.")
                continue
            
            # 读取日志文件
            try:
                ocel = ocel_import_factory.apply(file_path, parameters={
                    "obj_names": ["A", "O"],
                    "val_names": [],
                    "act_name": "event_activity",
                    "time_name": "event_timestamp",
                    "sep": ","
                })
            except Exception as e:
                print(f"Error reading file {file_path}: {e}")
                continue
            
            # 生成 Petri 网并计算执行时间
            start_time = time.time()
            try:
                ocpn = ocpn_discovery_factory.apply(ocel, parameters={"debug": False})
            except Exception as e:
                print(f"Error generating Petri net for file {file_path}: {e}")
                continue
            end_time = time.time()
            execution_time = end_time - start_time
            
            # 保存结果
            output_data.append({
                "文件名": file,
                "事件数": num_events,
                "执行时间": execution_time
            })
            
            print(f"Processed file {file} with {num_events} events. Execution time: {execution_time:.2f} seconds.")

# 将结果保存到 CSV 文件
output_df = pd.DataFrame(output_data)
output_df.to_csv(output_file, index=False, encoding="utf-8-sig")

print(f"Results saved to {output_file}")
```
